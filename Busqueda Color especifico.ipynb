```python
rr=[0,53]
gg=[15,78]
bb=[50,165]
```

Prints

```python
cube = io.imread('/home/cesar/Desktop/rb.jpg')
for x in range(cube.shape[0]):
    for y in range(cube.shape[1]):
        r=cube[x][y][0]
        g=cube[x][y][1]
        b=cube[x][y][2]
        
        if r>rr[0] and r<rr[1] and g>gg[0] and g<gg[1] and b>bb[0] and b<bb[1]:
            
            cube[x][y][0]=0
            cube[x][y][1]=0
            cube[x][y][2]=255
            
        else:
            cube[x][y][0]=255
            cube[x][y][1]=255
            cube[x][y][2]=255
            
io.imshow(cube)         
         
            
            
```

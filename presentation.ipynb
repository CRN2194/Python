
# Blur


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
img = cv2.imread('/home/cesar/Desktop/datasetmini/train/148.jpg')
b,g,r = cv2.split(img)      
imconv= cv2.merge([r,g,b])
blur = cv2.blur(imconv,(4,4))

plt.subplot(121),plt.imshow(imconv),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
plt.imsave("/home/cesar/Desktop/noise3.jpg",blur)

```

# MediamBlur

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
img = cv2.imread('/home/cesar/Downloads/ruido.jpg') 
b,g,r = cv2.split(img)      
imconv= cv2.merge([r,g,b])
median = cv2.medianBlur(imconv,9)
plt.subplot(121),plt.imshow(imconv),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imsave("/home/cesar/Desktop/error404.jpg",median)
```

# Bilatteral filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
img = cv2.imread('/home/cesar/Desktop/noise.jpg') 
b,g,r = cv2.split(img)      
imconv= cv2.merge([r,g,b])
bilateral = cv2.bilateralFilter(imconv,9,75,75)
plt.subplot(121),plt.imshow(imconv),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(bilateral),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imsave("/home/cesar/Desktop/noise2.jpg",bilateral)
```

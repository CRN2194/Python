# Hough Tranform in OpenCV


```python
import cv2
import numpy as np
from skimage import io
%pylab
%matplotlib inline
img = cv2.imread('/home/cesar/Desktop/dd.png')
io.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#print("gray",gray)
#print()
edges = cv2.Canny(gray,50,150,apertureSize = 3)
#print("edge,",edges)
#print()
io.imshow(edges)
```

```python
img = cv2.imread('/home/cesar/Desktop/dd.png')
lines = cv2.HoughLines(edges,1,np.pi/180,200)

#print("Lineas",lines)
for ln in lines:
    for rho,theta in ln:
        
        a = np.cos(theta)
        
        b = np.sin(theta)
        
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        

        cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)


io.imshow(img)
```

```python
import cv2
import numpy as np

img = cv2.imread('/home/cesar/Desktop/dd.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
minLineLength = 100
maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
for x1,y1,x2,y2 in lines[0]:
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)

cv2.imwrite('houghlines5.jpg',img)
io.imshow(img)
```
